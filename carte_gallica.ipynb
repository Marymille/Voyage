```python
script
import requests
from bs4 import BeautifulSoup
import json

# Liste des images et leurs URLs sur Gallica
cartes = [
    {
        "image": "1635.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b53143185p.r=cor%C3%A9e%20du%20sud?rk=321890;0"
    },
    {
        "image": "1652-1656.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b52522048w.r=cor%C3%A9e%20du%20sud?rk=171674;4"
    },
    {
        "image": "1737.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b53027431f.r=cor%C3%A9e%20du%20sud?rk=42918;4"
    },
    {
        "image": "~1750.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b53069751b.r=cor%C3%A9e%20du%20sud?rk=278971;2"
    },
    {
        "image": "1785.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b55000442r.r=cor%C3%A9e%20du%20sud?rk=42918;4"
    },
    {
        "image": "1883.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b530948152/f1.item.r=cor%C3%A9e%20du%20sud"
    },
    {
        "image": "1894.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b8440820p.r=cor%C3%A9e%20du%20sud?rk=21459;2"
    },
    {
        "image": "1894-1.jpg",
        "url": "https://gallica.bnf.fr/ark:/12148/btv1b531671095.r=cor%C3%A9e%20du%20sud?rk=193134;0"
    }
]

# Fonction pour extraire les métadonnées d'une page Gallica
def get_metadata(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extraction des métadonnées
    metadata = {}
    
    try:
        # Récupération du titre
        title_tag = soup.find('meta', {'name': 'DC.title'})
        metadata['title'] = title_tag['content'] if title_tag else 'Titre non disponible'
        
        # Récupération de l'auteur
        author_tag = soup.find('meta', {'name': 'DC.creator'})
        metadata['author'] = author_tag['content'] if author_tag else 'Auteur non disponible'
        
        # Récupération de l'éditeur
        publisher_tag = soup.find('meta', {'name': 'DC.publisher'})
        metadata['publisher'] = publisher_tag['content'] if publisher_tag else 'Éditeur non disponible'
        
        # Récupération de la date
        date_tag = soup.find('meta', {'name': 'DC.date'})
        metadata['date'] = date_tag['content'] if date_tag else 'Date non disponible'
        
        # Récupération des sujets
        subject_tags = soup.find_all('meta', {'name': 'DC.subject'})
        metadata['subject'] = ', '.join([subject['content'] for subject in subject_tags]) if subject_tags else 'Sujets non disponibles'
        
        # Lien vers la notice complète
        metadata['link'] = url
        
        # Description
        description_tag = soup.find('meta', {'name': 'description'})
        metadata['description'] = description_tag['content'] if description_tag else 'Description non disponible'
        
        # Format et échelle
        metadata['type'] = 'Carte'
        metadata['language'] = 'Français'  # À ajuster si nécessaire
        metadata['scale'] = '1:4 500 000'  # À ajuster si nécessaire
        
    except Exception as e:
        print(f"Erreur lors de l'extraction des métadonnées pour l'URL {url}: {e}")
    
    return metadata

# Liste pour stocker les informations des cartes
cartes_metadata = []

# Récupérer les métadonnées pour chaque carte
for carte in cartes:
    print(f"Récupération des métadonnées pour {carte['image']}...")
    metadata = get_metadata(carte['url'])
    metadata['image'] = carte['image']  # Associer l'image avec les métadonnées
    cartes_metadata.append(metadata)

# Sauvegarder les données dans un fichier JSON
with open('cartes_metadata.json', 'w', encoding='utf-8') as f:
    json.dump(cartes_metadata, f, ensure_ascii=False, indent=4)

print("Les métadonnées des cartes ont été sauvegardées dans le fichier cartes_metadata.json.")
```
